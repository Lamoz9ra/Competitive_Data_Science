```python
import pandas as pd
```

```python
df = pd.read_csv('../Course Data/data/car_info.csv')
```

```python
df.head(5)
```

```python
df = df.sort_values(by=['year_to_start','car_rating'], ascending=[False, True])
df.head(5)
```

```python
cols2drop = ['car_id','car']
```

```python
cols2drop = [col for col in cols2drop if col in df.columns]
df = df.drop(columns=cols2drop)
```

```python
df.head(5)
```

```python
columns_order_list = df.columns[:-1]
columns_order_list
```

```python
columns_order_list = [col for col in columns_order_list if col in df.columns]
df = df[columns_order_list]
df.head(5)
```

```python
critical_car_rating = 4
```

```python
df = df[df["car_rating"] >= critical_car_rating]
df.head(5)
```
